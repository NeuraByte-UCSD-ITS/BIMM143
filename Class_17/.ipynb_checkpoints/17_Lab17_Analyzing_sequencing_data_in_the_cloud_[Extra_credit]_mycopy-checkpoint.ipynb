````r
# Load required libraries
library(tximport)
library(DESeq2)
library(rhdf5)  # Add this line to load the rhdf5 package

# Set up the folder and filenames to read
setwd("/Users/edwinruiz/ComputerScience/BIMM143/BIMM143/BIMM143/Class_17")

folders <- dir(pattern="SRR21568*")
samples <- sub("_quant", "", folders)
files <- file.path(folders, "abundance.h5")
names(files) <- samples

# Import Kallisto results
txi.kallisto <- tximport(files, type = "kallisto", txOut = TRUE)

# Display the counts
head(txi.kallisto$counts)

# Summarize the total counts per sample
colSums(txi.kallisto$counts)

# Determine the number of transcripts detected in at least one sample
sum(rowSums(txi.kallisto$counts) > 0)

# Filter out transcripts with no reads
to.keep <- rowSums(txi.kallisto$counts) > 0
kset.nonzero <- txi.kallisto$counts[to.keep,]

# Filter out transcripts with no change over the samples
keep2 <- apply(kset.nonzero, 1, sd) > 0
x <- kset.nonzero[keep2,]

# Perform PCA
pca <- prcomp(t(x), scale = TRUE)
summary(pca)

# Plot PCA
plot(pca$x[, 1], pca$x[, 2], col = c("blue", "blue", "red", "red"), xlab = "PC1", ylab = "PC2", pch = 16)

# Use ggplot to make a similar figure of PC1 vs PC2 and a separate figure PC1 vs PC3 and PC2 vs PC3
library(ggplot2)

# Create a data frame for ggplot
pca_data <- data.frame(Sample = rownames(pca$x), PC1 = pca$x[, 1], PC2 = pca$x[, 2], PC3 = pca$x[, 3])
pca_data$Condition <- c("Control", "Control", "Treatment", "Treatment")

# Plot PC1 vs PC2
ggplot(pca_data, aes(x = PC1, y = PC2, color = Condition)) +
    geom_point(size = 4) +
    labs(title = "PCA - PC1 vs PC2", x = "PC1", y = "PC2") +
    theme_minimal()

# Plot PC1 vs PC3
ggplot(pca_data, aes(x = PC1, y = PC3, color = Condition)) +
    geom_point(size = 4) +
    labs(title = "PCA - PC1 vs PC3", x = "PC1", y = "PC3") +
    theme_minimal()

# Plot PC2 vs PC3
ggplot(pca_data, aes(x = PC2, y = PC3, color = Condition)) +
    geom_point(size = 4) +
    labs(title = "PCA - PC2 vs PC3", x = "PC2", y = "PC3") +
    theme_minimal()

# Differential expression analysis
sampleTable <- data.frame(condition = factor(rep(c("control", "treatment"), each = 2)))
rownames(sampleTable) <- colnames(txi.kallisto$counts)

dds <- DESeqDataSetFromTximport(txi.kallisto, sampleTable, ~condition)

dds <- DESeq(dds)

res <- results(dds)
head(res)
````

````r
# Load required libraries
library(tximport)
library(DESeq2)
library(rhdf5)
library(ggplot2)
library(ggrepel)

# Set up the folder and filenames to read
setwd("/Users/edwinruiz/ComputerScience/BIMM143/BIMM143/BIMM143/Class_17")

folders <- dir(pattern="SRR21568*")
samples <- sub("_quant", "", folders)
files <- file.path(folders, "abundance.h5")
names(files) <- samples

# Import Kallisto results
txi.kallisto <- tximport(files, type = "kallisto", txOut = TRUE)

# Display the counts
head(txi.kallisto$counts)

# Summarize the total counts per sample
colSums(txi.kallisto$counts)

# Determine the number of transcripts detected in at least one sample
sum(rowSums(txi.kallisto$counts) > 0)

# Filter out transcripts with no reads
to.keep <- rowSums(txi.kallisto$counts) > 0
kset.nonzero <- txi.kallisto$counts[to.keep,]

# Filter out transcripts with no change over the samples
keep2 <- apply(kset.nonzero, 1, sd) > 0
x <- kset.nonzero[keep2,]

# Perform PCA
pca <- prcomp(t(x), scale = TRUE)
summary(pca)

plot(pca$x[,1], pca$x[,2],
     col=c("blue","blue","red","red"),
     xlab="PC1", ylab="PC2", pch=16)

# Plot PCA using ggplot
pca_data <- data.frame(Sample = rownames(pca$x), PC1 = pca$x[, 1], PC2 = pca$x[, 2], PC3 = pca$x[, 3])
pca_data$Condition <- c("Control", "Control", "Treatment", "Treatment")

# Make metadata object for the samples
colData <- data.frame(condition = factor(rep(c("control", "treatment"), each = 2)))
rownames(colData) <- colnames(txi.kallisto$counts)

# Make the data frame for ggplot 
y <- as.data.frame(pca$x)
y$Condition <- as.factor(colData$condition)

ggplot(y) +
  aes(PC1, PC2, col=Condition) +
  geom_point() +
  geom_text_repel(label=rownames(y)) +
  theme_bw()

#PC1 vs PC2
ggplot(pca_data, aes(x = PC1, y = PC2, color = Condition)) +
    geom_point(size = 4) +
    labs(title = "PCA - PC1 vs PC2", x = "PC1", y = "PC2") +
    theme_minimal()

# Plot PC1 vs PC3
ggplot(pca_data, aes(x = PC1, y = PC3, color = Condition)) +
    geom_point(size = 4) +
    labs(title = "PCA - PC1 vs PC3", x = "PC1", y = "PC3") +
    theme_minimal()

# Plot PC2 vs PC3
ggplot(pca_data, aes(x = PC2, y = PC3, color = Condition)) +
    geom_point(size = 4) +
    labs(title = "PCA - PC2 vs PC3", x = "PC2", y = "PC3") +
    theme_minimal()

# Differential expression analysis
sampleTable <- data.frame(condition = factor(rep(c("control", "treatment"), each = 2)))
rownames(sampleTable) <- colnames(txi.kallisto$counts)

dds <- DESeqDataSetFromTximport(txi.kallisto, sampleTable, ~condition)

dds <- DESeq(dds)

res <- results(dds)
head(res)
````
